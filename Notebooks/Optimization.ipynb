# Optimization Inspection

```python
import numpy as np
import pandas as pd

PSO_LOG = pd.DataFrame(np.genfromtxt('../log/PSO_FLVL'))
PSO_LOG.columns = ['rmse', 'hidden_size', 'dropout', 'num_layers', 'num_timesteps', 'lr_init', 'gamma', 'weight_decay']
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 4, figsize = (10,8), dpi = 300)

# hidden_size
ax[0][0].scatter(PSO_LOG['hidden_size'], PSO_LOG['rmse'], c = PSO_LOG.index, alpha = 0.7)
ax[0][0].set_yscale('log')
ax[0][0].set_xlabel('hidden_size')
ax[0][0].set_ylabel('Root Mean Squared Error (RMSE)')

# dropout
ax[0][1].scatter(PSO_LOG['dropout'], PSO_LOG['rmse'], c = PSO_LOG.index, alpha = 0.7)
ax[0][1].set_yscale('log')
ax[0][1].set_xlabel('droupout')

# num_layers
ax[0][2].scatter(PSO_LOG['num_layers'], PSO_LOG['rmse'], c = PSO_LOG.index, alpha = 0.7)
ax[0][2].set_yscale('log')
ax[0][2].set_xlabel('num_layers')

# num_timesteps
ax[0][3].scatter(PSO_LOG['num_timesteps'], PSO_LOG['rmse'], c = PSO_LOG.index, alpha = 0.7)
ax[0][3].set_yscale('log')
ax[0][3].set_xlabel('num_timesteps')

# lr_init
ax[1][0].scatter(PSO_LOG['lr_init'], PSO_LOG['rmse'], c = PSO_LOG.index, alpha = 0.7)
ax[1][0].set_yscale('log')
ax[1][0].set_xscale('log')
ax[1][0].set_xlabel('lr_init')
ax[1][0].set_ylabel('Root Mean Squared Error (RMSE)')

# gamma
ax[1][1].scatter(PSO_LOG['gamma'], PSO_LOG['rmse'], c = PSO_LOG.index, alpha = 0.7)
ax[1][1].set_yscale('log')
ax[1][1].set_xlabel('gamma')

# weight_decay
ax[1][2].scatter(PSO_LOG['weight_decay'], PSO_LOG['rmse'], c = PSO_LOG.index, alpha = 0.7)
ax[1][2].set_yscale('log')
ax[1][2].set_xlabel('weight_decay')

# Colorbar
cmap = plt.cm.viridis
for i in np.linspace(PSO_LOG.index.min(), PSO_LOG.index.max(), 1000):
    ax[1][3].hlines(i, 0, 1, color = cmap(i / PSO_LOG.index.max()), alpha = 0.7)
ax[1][3].set_xlim(0, 1)
ax[1][3].set_ylim(0, PSO_LOG.index.max())
ax[1][3].set_xticks(ticks = [])
ax[1][3].set_ylabel("Interation")

plt.tight_layout()
plt.show()
```
